```python
from tkinter import *
from tkinter import ttk

from tkinter import messagebox


class StudentManagementSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("Student Management System")
        self.root.geometry("1540x800+0+0")

        # Title
        lbltitle = Label(self.root, text="STUDENT MANAGEMENT SYSTEM", bd=20, relief=RIDGE, font=("times new roman", 50, "bold"), fg="red", bg="white")
        lbltitle.pack(side=TOP, fill=X)

        # Variables
        self.NameOfStudent = StringVar()
        self.ReferenceNo = StringVar()
        self.RollNo = StringVar()
        self.Class = StringVar()
        self.Section = StringVar()
        self.DOB = StringVar()
        self.Phone = StringVar()
        self.Address = StringVar()
        self.GuardianName = StringVar()
        self.GuardianPhone = StringVar()
        self.Email = StringVar()
        self.ParentOccupation = StringVar()

        # List of students for the dropdown (ComboBox)
        self.student_names = ["anchal rai", "ram tyagi", "sita", "sunil rai", "radha kumari","anjana","harsh"]

        # Frames
        DataFrame = Frame(self.root, bd=20, relief=RIDGE)
        DataFrame.place(x=0, y=130, width=1530, height=400)

        DataFrameLeft = LabelFrame(DataFrame, bd=10, relief=RIDGE, padx=10, font=("arial", 12, "bold"), text="Student Information")
        DataFrameLeft.place(x=0, y=5, width=980, height=350)

        DataFrameRight = LabelFrame(DataFrame, bd=10, relief=RIDGE, padx=10, font=("arial", 12, "bold"), text="Further Details")
        DataFrameRight.place(x=990, y=5, width=460, height=350)

        ButtonFrame = Frame(self.root, bd=20, relief=RIDGE)
        ButtonFrame.place(x=0, y=530, width=1530, height=70)

        DetailsFrame = Frame(self.root, bd=20, relief=RIDGE)
        DetailsFrame.place(x=0, y=600, width=1530, height=170)

        # DataFrameLeft (Input Fields)
        lblNameOfStudent = Label(DataFrameLeft, text="Name of Student:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblNameOfStudent.grid(row=0, column=0, sticky=W)

        # Dropdown for selecting student name
        cmbNameOfStudent = ttk.Combobox(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.NameOfStudent, width=29)
        cmbNameOfStudent['values'] = self.student_names
        cmbNameOfStudent.grid(row=0, column=1)

        lblReferenceNo = Label(DataFrameLeft, text="Reference No:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblReferenceNo.grid(row=1, column=0, sticky=W)
        txtReferenceNo = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.ReferenceNo, width=29)
        txtReferenceNo.grid(row=1, column=1)

        lblRollNo = Label(DataFrameLeft, text="Roll No:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblRollNo.grid(row=2, column=0, sticky=W)
        txtRollNo = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.RollNo, width=29)
        txtRollNo.grid(row=2, column=1)

        lblClass = Label(DataFrameLeft, text="Class:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblClass.grid(row=3, column=0, sticky=W)
        txtClass = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.Class, width=29)
        txtClass.grid(row=3, column=1)

        lblSection = Label(DataFrameLeft, text="Section:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblSection.grid(row=4, column=0, sticky=W)
        txtSection = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.Section, width=29)
        txtSection.grid(row=4, column=1)

        lblDOB = Label(DataFrameLeft, text="Date of Birth:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblDOB.grid(row=5, column=0, sticky=W)
        txtDOB = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.DOB, width=29)
        txtDOB.grid(row=5, column=1)

        lblPhone = Label(DataFrameLeft, text="Phone Number:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblPhone.grid(row=6, column=0, sticky=W)
        txtPhone = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.Phone, width=29)
        txtPhone.grid(row=6, column=1)

        lblAddress = Label(DataFrameLeft, text="Address:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblAddress.grid(row=7, column=0, sticky=W)
        txtAddress = Entry(DataFrameLeft, font=("arial", 12, "bold"), textvariable=self.Address, width=29)
        txtAddress.grid(row=7, column=1)

        # DataFrameRight (Further Details)
        lblGuardianName = Label(DataFrameRight, text="Guardian Name:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblGuardianName.grid(row=0, column=0, sticky=W)
        txtGuardianName = Entry(DataFrameRight, font=("arial", 12, "bold"), textvariable=self.GuardianName, width=29)
        txtGuardianName.grid(row=0, column=1)

        lblGuardianPhone = Label(DataFrameRight, text="Guardian Phone:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblGuardianPhone.grid(row=1, column=0, sticky=W)
        txtGuardianPhone = Entry(DataFrameRight, font=("arial", 12, "bold"), textvariable=self.GuardianPhone, width=29)
        txtGuardianPhone.grid(row=1, column=1)

        lblEmail = Label(DataFrameRight, text="Email:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblEmail.grid(row=2, column=0, sticky=W)
        txtEmail = Entry(DataFrameRight, font=("arial", 12, "bold"), textvariable=self.Email, width=29)
        txtEmail.grid(row=2, column=1)

        lblParentOccupation = Label(DataFrameRight, text="Parent Occupation:", font=("arial", 12, "bold"), padx=2, pady=6)
        lblParentOccupation.grid(row=3, column=0, sticky=W)
        txtParentOccupation = Entry(DataFrameRight, font=("arial", 12, "bold"), textvariable=self.ParentOccupation, width=29)
        txtParentOccupation.grid(row=3, column=1)

        # Buttons
        btnAdd = Button(ButtonFrame, text="Add", font=("arial", 12, "bold"), width=20, bg="green", fg="white", command=self.add_student)
        btnAdd.grid(row=0, column=0)

        btnUpdate = Button(ButtonFrame, text="Update", font=("arial", 12, "bold"), width=20, bg="green", fg="white", command=self.update_student)
        btnUpdate.grid(row=0, column=1)

        btnDelete = Button(ButtonFrame, text="Delete", font=("arial", 12, "bold"), width=20, bg="green", fg="white", command=self.delete_student)
        btnDelete.grid(row=0, column=2)

        btnClear = Button(ButtonFrame, text="Clear", font=("arial", 12, "bold"), width=20, bg="green", fg="white", command=self.clear_fields)
        btnClear.grid(row=0, column=3)

        btnExit = Button(ButtonFrame, text="Exit", font=("arial", 12, "bold"), width=20, bg="green", fg="white", command=self.confirm_exit)
        btnExit.grid(row=0, column=4)

        # Details Frame (Table)
        scroll_x = Scrollbar(DetailsFrame, orient=HORIZONTAL)
        scroll_y = Scrollbar(DetailsFrame, orient=VERTICAL)
        self.student_table = ttk.Treeview(
            DetailsFrame,
            columns=("Name", "Reference", "RollNo", "Class", "Section", "DOB", "Phone", "Address", "GuardianName", "GuardianPhone", "Email", "ParentOccupation"),
            xscrollcommand=scroll_x.set,
            yscrollcommand=scroll_y.set,
        )
        scroll_x.pack(side=BOTTOM, fill=X)
        scroll_y.pack(side=RIGHT, fill=Y)
        scroll_x.config(command=self.student_table.xview)
        scroll_y.config(command=self.student_table.yview)

        self.student_table.heading("Name", text="Name")
        self.student_table.heading("Reference", text="Reference No")
        self.student_table.heading("RollNo", text="Roll No")
        self.student_table.heading("Class", text="Class")
        self.student_table.heading("Section", text="Section")
        self.student_table.heading("DOB", text="Date of Birth")
        self.student_table.heading("Phone", text="Phone Number")
        self.student_table.heading("Address", text="Address")
        self.student_table.heading("GuardianName", text="Guardian Name")
        self.student_table.heading("GuardianPhone", text="Guardian Phone")
        self.student_table.heading("Email", text="Email")
        self.student_table.heading("ParentOccupation", text="Parent Occupation")
        self.student_table["show"] = "headings"

        self.student_table.pack(fill=BOTH, expand=1)

    def add_student(self):
        # Add student to the Treeview (or a database)
        self.student_table.insert("", END, values=(
            self.NameOfStudent.get(),
            self.ReferenceNo.get(),
            self.RollNo.get(),
            self.Class.get(),
            self.Section.get(),
            self.DOB.get(),
            self.Phone.get(),
            self.Address.get(),
            self.GuardianName.get(),
            self.GuardianPhone.get(),
            self.Email.get(),
            self.ParentOccupation.get(),
        ))

    def update_student(self):
        # Code to update student information (e.g., based on Reference No or other identifier)
        pass

    def delete_student(self):
        selected_item = self.student_table.selection()
        if selected_item:
            self.student_table.delete(selected_item)

    def clear_fields(self):
        # Clear all input fields
        self.NameOfStudent.set("")
        self.ReferenceNo.set("")
        self.RollNo.set("")
        self.Class.set("")
        self.Section.set("")
        self.DOB.set("")
        self.Phone.set("")
        self.Address.set("")
        self.GuardianName.set("")
        self.GuardianPhone.set("")
        self.Email.set("")
        self.ParentOccupation.set("")

    def confirm_exit(self):
        # Ask for confirmation when the user tries to exit
        result = messagebox.askquestion("Exit", "Are you sure you want to exit?")
        if result == 'yes':
            self.root.destroy()

# Main Function
if __name__ == "__main__":
    root = Tk()
    app = StudentManagementSystem(root)
    root.mainloop()
```
